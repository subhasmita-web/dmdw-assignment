```python
import matplotlib.pyplot as plt
import numpy as np

#simple bar and scatter plot

#assume there are 5 students

x=np.arange(5)

#their test scores

y=(20,35,30,35,27)

#bar plot

plt.bar(x,y)
plt.show()

#scatter plot

plt.scatter(x,y)
plt.show()

#box plot
plt.boxplot(x,y)
plt.show()

#histogram(data is inappropriate here we must need a relationship between 2 points to draw this )
plt.hist(x,y)
plt.show()

#line plot
plt.line(x,y)
plt.show()
```

```python
'''plt.bar=creates a bar chart
plt.scatter=makes a scatter plot
plt.boxplot=makes a box plot
plt.hist=makes a histogram
plt.plot=creates a lineplot'''
```

```python
#customizing Labels
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#generate sample data
#100-evenly-spaced values from 0 to 20
x=np.linspace(0,20,100)
print(x)
y=np.sin(x)
print(y)

plt.plot(x,y,label='Sample Label')
plt.show()

#customize axis labels
plt.plot(x,y,label='sample label')

#x axis title
plt.xlabel("x axis label")

#y axis title
plt.ylabel("y axis label")

#chart title
plt.title("Sample Plot Title")
plt.show()

#Saving chart to file
plt.savefig("new_chart.jpg")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#generate sample data

x=np.linspace(0,20,100)
y=np.sin(x)

#get an empty figure and add an axes
fig=plt.figure(figsize=(8,4))
#row-column
ax=fig.add_subplot(1,1,1)
#line plot data on the axes
ax.plot(x,y)
ax.plot(x,y,'b-',linewidth=2,label='Sample Label')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#generate sample dara for multiple line
x1=np.linspace(0,100,20)
x2=np.linspace(0,100,20)
x3=np.linspace(0,100,20)
y1=np.sin(x1)
y2=np.cos(x2)
y3=np.tan(x3)

#get the figure and axes all at once using subplots
fig,ax=plt.subplots(figsize=(8,4))

#line plot data on the axes
ax.plot(x1,y1,label='sin')
ax.plot(x2,y2,label='cos')
ax.plot(x3,y3,label='tan')

#add legend
ax.legend(loc='best',prop={'size':'large'})

#add grid and title
ax1.grid(True)
ax2.grid(True)
ax3.grid(True)
```
